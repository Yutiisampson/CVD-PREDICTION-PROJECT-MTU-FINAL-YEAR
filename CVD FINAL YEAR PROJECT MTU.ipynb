# Outlier Detection and Removal

```python
lst = [ 'Height_(cm)', 'Weight_(kg)', 'BMI','Alcohol_Consumption', 'Fruit_Consumption','Green_Vegetables_Consumption', 'FriedPotato_Consumption']
for i in lst:
  plt.figure(figsize=(3,3))
  sb.boxplot(x=i,data=cardio,showmeans=True)
  plt.show()
```

The boxplot, along with the code, clearly indicates the presence of outliers. When comparing the mean and maximum values, it is evident that the maximum value is significantly higher than the mean.

```python
# Outlier removal
numerical_cols = ['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption', 'Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']
for col in numerical_cols:
    Q1 = cardio[col].quantile(0.25)
    Q3 = cardio[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    cardio = cardio[(cardio[col] >= lower) & (cardio[col] <= upper)]
cardio.reset_index(drop=True, inplace=True)
print("Shape after outlier removal:", cardio.shape)
```

This section identifies and removes outliers in numerical columns using the Interquartile Range (IQR) method to improve model performance.

- **Boxplots**: Visualizes numerical columns (`Height_(cm)`, `Weight_(kg)`, `BMI`, etc.) to detect outliers.
- **IQR Method**: For each numerical column:
  - Computes Q1 (25th percentile), Q3 (75th percentile), and IQR.
  - Removes rows where values fall outside `[Q1 - 1.5*IQR, Q3 + 1.5*IQR]`.
- **Result**: Reduces dataset size to `185737 rows, 19 columns` after outlier removal.

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
print("\nComputing Mutual Information...")
mi_scores = mutual_info_classif(X_processed, y, random_state=42)
mi_scores = pd.Series(mi_scores, index=X_processed.columns, name='MI_Score')
```

The result is stored in a **Series** indexed by feature names, representing how informative each feature is

The features are sorted in descending order of their mutual information score.

The top 10 features with the highest MI scores are displayed and the features **Exercise_1, Checkup_In the past year, and Gender_Male** are among the most informative.
